```python
def sex(fun):
    if fun=='F':
        return 0
    else:
        return 1
```

```python
train['sex']=train['SEX'].apply(sex)
```

```python
train_1=train.drop(['patient_id_','SEX','SOURCE'],axis=1)
```

```python
test_1=test[train_1.columns]
```

```python
from sklearn.model_selection import train_test_split
train_test_split
```

```python
train_1.head()
```

```python
X=train_1
y=train.SOURCE
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=2)
```

```python
from sklearn.ensemble import RandomForestClassifier
random=RandomForestClassifier(n_estimators=1000,max_depth=100)
```

```python
random.fit(X_train,y_train)
```

```python
ran_predict=random.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix,accuracy_score
```

```python
print(classification_report(y_test,ran_predict))
```

```python
print(confusion_matrix(y_test,ran_predict))
```

```python
sns.heatmap(confusion_matrix(y_test,ran_predict,),annot=True)
submit=pd.read_csv('sample_submission_AI.csv')
```
